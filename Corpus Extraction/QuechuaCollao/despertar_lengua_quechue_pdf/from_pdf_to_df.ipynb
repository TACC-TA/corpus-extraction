```python
import pdfplumber
import csv

# Open the PDF
with pdfplumber.open('/content/El despertar de mi lengua quechua_ El entramado de mi waraka.pdf') as pdf:
    with open('output.csv', 'w', newline='', encoding='utf-8-sig') as csvfile:
        writer = csv.writer(csvfile)

        # Iterate through pages
        for page in pdf.pages:
            text = page.extract_text()
            if text:  # Check if text is not None
                # Process text into lines and columns
                for line in text.split('\n'):
                    # Assuming data is separated by commas in the text
                    row = line.split(',')
                    writer.writerow(row)
```

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

def remove_unwanted_words(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in lines]
    return ' '.join(filtered_words)
```

```python
import string
from nltk.tokenize import word_tokenize

def remove_punctuation(text):
    # Create a translation table that maps each punctuation character to None
    translator = str.maketrans('', '', string.punctuation)
    # Use translate to remove punctuation
    return text.translate(translator)

# Apply the function to the DataFrame
df_['No_Punctuation'] = df_['Filtered_Text'].apply(remove_punctuation)

# Print the result
df_.head(50)
```

```python
import re
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

# Apply the function to the DataFrame
df_['No_Numbers'] = df_['No_Punctuation'].apply(remove_numbers)

# Print the result
df_.head(50)
```

```python
dff = pd.DataFrame(columns=['Text'])
dff['Text'] = df_['No_Numbers']
dff.head()
```

```python
dff.replace('', pd.NA, inplace=True)  # Replace empty strings with NaN
dff.dropna(how='all', inplace=True)   # Then drop the empty rows
dff.head(50)
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')

# Load Spanish stopwords
spanish_stopwords = stopwords.words('spanish')

def remove_stopwords(text):
    words = word_tokenize(text, language='spanish')
    filtered_words = [word for word in words if word.lower() not in spanish_stopwords]
    return ' '.join(filtered_words)

dff['Cleaned_Sentences'] = dff['Text'].apply(remove_stopwords)

# Display the result
dff.head(50)
```

```python
df = pd.DataFrame()
```

```python
df['Text'] = dff['Cleaned_Sentences']
```

```python
df['Text'].replace('', pd.NA, inplace=True)  # Replace empty strings with NaN
df['Text'].dropna(how='all', inplace=True)   # Then drop the empty rows
df['Text'] = df['Text'].reset_index(drop=True)
df['Text'] = df['Text'].iloc[:659]
df['Text'].tail(50)
```

```python
df['Text'] = df['Text'].str.lower()
```

```python
df['Text'].head(50)
```

```python
# Remove standalone letters
df['Cleaned_Sentences'] = df['Text'].str.replace(r'\b[a-zA-Z]\b', '', regex=True)

# Optional: remove any extra spaces left behind
df['Cleaned_Sentences'] = df['Cleaned_Sentences'].str.replace(r'\s{2,}', ' ', regex=True).str.strip()
```
